```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

files = {
    "QB": "QBDataCube_blended.csv",
    "Rushing": "RushingDataCube_blended.csv",
    "Receiving": "ReceivingDataCube_blended.csv",
    "Defense": "DefenseDataCube_blended.csv",
    "Kicking": "KickingDataCube_blended.csv"
}

all_data = []

for position, filename in files.items():
    df = pd.read_csv(filename)
    name_col = df.columns[0]
    career_lengths = df.groupby(name_col).size()
    for length in career_lengths:
        all_data.append({"Position": position, "Career Length": length})

career_df = pd.DataFrame(all_data)

plt.figure(figsize=(12,8))
sns.violinplot(data=career_df, x="Position", y="Career Length", inner="box", cut=0)
plt.title("Distribution of NFL Career Lengths by Position", fontsize=18)
plt.xlabel("Position Group", fontsize=14)
plt.ylabel("Career Length (Seasons)", fontsize=14)
plt.tight_layout()
plt.show()
```
