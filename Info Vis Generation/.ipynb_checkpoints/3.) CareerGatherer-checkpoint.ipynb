```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random

# Load the data
df = pd.read_csv("QBDataCube_blended.csv")

name_col = df.columns[0]
score_col = df.columns[-1]

# Choose 100 random players
sample_players = (
    df[name_col]
    .drop_duplicates()
    .sample(n=100)
)

# Create the sample data frame
sample_df = df[df[name_col].isin(sample_players)]

# Gather all XKCD colors
xkcd_colors = list(mcolors.XKCD_COLORS.values())

# Now graph it all!

plt.figure(figsize=(14, 8))
for player_name, player_data in sample_df.groupby(name_col):
    scores = player_data[score_col].values
    career_years = range(len(scores))
    plt.plot(career_years, scores, color=random.choice(xkcd_colors), alpha=1, linewidth=3)
plt.title("Career Trajectories of 100 Random NFL Quarterbacks", fontsize=18)
plt.xlabel("Num Years Into NFL Career", fontsize=14)
plt.ylabel("Percentile Score", fontsize=14)
plt.grid(True)
plt.tight_layout()

plt.show()
```
