```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

df = pd.read_csv("QBDataCube_blended.csv")

# Gather the correct columns (double check this each time we alter this code!)
passing_yards_col = df.columns[2]
percentile_col = df.columns[-1]

df[passing_yards_col] = pd.to_numeric(df[passing_yards_col])
df[percentile_col] = pd.to_numeric(df[percentile_col])
correlation, p_value = pearsonr(df[passing_yards_col], df[percentile_col])

print(f"Correlation = {correlation:.4f}")
print(f"P-value = {p_value:.4e}")

# Now let's plot it all!

plt.figure(figsize=(12,8))

sns.regplot(data=df, x=passing_yards_col, y=percentile_col,
            scatter_kws={"color": "forestgreen", "alpha": 0.40, "s": 18},
            line_kws={"color": "black", "linewidth": 3})
plt.plot([], [], ' ', label=(f"Pearson r = {correlation:.3f}\n"
                             f"p-value = {p_value:.2e}"))
plt.legend(loc="best", fontsize=12)
plt.title("Quarterback Passing Yards vs Percentile Score", fontsize=18)
plt.xlabel("Passing Yards", fontsize=14)
plt.ylabel("Percentile Score", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
